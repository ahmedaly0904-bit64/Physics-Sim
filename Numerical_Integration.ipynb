```python
import numpy as np
def f(x):
    return x**4-2*x+1
```

```python
def trapziodal_integration(a,b,N):
    h = (b-a)/N
    x = np.linspace(a,b,N+1)
    y = f(x)
    s = y[0] + y[-1] + 2 * np.sum(y[1:-1])
    result = (h/2)*s
    return result
I = trapziodal_integration(0,2,100)
print(f"Claculted Area (N=10) = {I}")
print("Actual Area = 4.4")
```
