```python
import matplotlib.pyplot as plt
import numpy as np
g =9.81
v0 = 50
t = np.linspace(0,10,100)
y = v0*t-0.5*g*t**2

plt.figure(figsize=(8,5))
plt.plot(t,y ,color='g')
plt.title("Projectile")
plt.xlabel("Time")
plt.ylabel("Height")
plt.axhline(0,color='b',linewidth=1)
plt.grid(True)
plt.show()
```
