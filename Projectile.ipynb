## 1. Basic Projectile Motion (Earth)
Simulation of a single projectile under standard gravity.

```python
import matplotlib.pyplot as plt
import numpy as np
# 1. Physical Constants
g =9.81
v0 = 50 # Initial Velocity in m/s
# 2. Time Setup
t = np.linspace(0,10,100)
# 3. Equation of motion
y = v0*t-0.5*g*t**2
# 4. Visualization & Simulation Setup
plt.figure(figsize=(8,5))
plt.plot(t,y ,color='g')
plt.title("Projectile")
plt.xlabel("Time")
plt.ylabel("Height")
plt.axhline(0,color='b',linewidth=1)
plt.grid(True)
plt.show()
```

## 2. Interplanetary Projectile Motion
In this section, we compare the trajectory of a projectile launched with the same velocity ($v_0 = 50 m/s$) on three different celestial bodies:
* **Earth** ($g \approx 9.81 m/s^2$)
* **Moon** ($g \approx 1.62 m/s^2$)
* **Jupiter** ($g \approx 24.79 m/s^2$)

```python
import numpy as np
import matplotlib.pyplot as plt
# 1.Physical constants
Planets = {
    "Earth":9.81,
    "Moon":1.62,
    "Jupiter":24.79
} 
v0 = 50 # Initial Velocity in m/s
# 2. Time setup
# We use 100 seconds to allow the projectile on the Moon to hit the ground
t = np.linspace(0,100,1000)

plt.figure(figsize=(8,5))

# 3. Physics Simulations Loop
for Planet, g in Planets.items():
    # Equation of motion
    y = v0*t-0.5*g*t**2
    plt.plot(t , y , label = Planet)
# 4. Visualization Setup
plt.ylim(0, 800)  
plt.title("Projectile Motion : Earth vs Moon vs Jupiter")
plt.xlabel("Time")
plt.ylabel("Height")
plt.grid(True)
plt.legend()
plt.show()
```
