```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Independent variable: growth rate r
# Increased number of points for higher resolution
r = np.linspace(0.1, 3.9, 1000) 

# 2. Initial population state
x = 0.5 * np.ones_like(r)  # array of same size as r filled with 0.5

# 3. Iteration loop (simulation)
# Let the system run for a transient period to settle
iterations = 100
for i in range(iterations):
    x = r * x * (1 - x)

# 4. Plotting
# Now x contains final values after 100 iterations
# Independent variable r (x-axis), dependent variable x (y-axis)
plt.plot(r, x, ',k', alpha=0.5) # ',k' draws very small black points

# 5. Set limits (focus on data region only)
plt.xlim(0, 4)   # r limits
plt.ylim(0, 1)   # x limits

plt.title("Logistic Map: Bifurcation Diagram")
plt.xlabel("Growth Rate (r)")
plt.ylabel("Population (x)")
plt.grid(True)
plt.show()
```
