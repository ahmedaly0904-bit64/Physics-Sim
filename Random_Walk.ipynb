```python
import numpy as np
import matplotlib.pyplot as plt
# make random values in an array (orders of steps)
steps = 100000
dx = np.random.choice([-1,1],steps)
dy = np.random.choice([-1,1],steps)
# coordinates of x, y
x = np.concatenate([[0], np.cumsum(dx)])
y = np.concatenate([[0], np.cumsum(dy)])

plt.figure(figsize=(8,8))
plt.plot(x, y, label='Random Path', linewidth=0.8, color='blue', alpha=0.7)

plt.title("2D Random Walk (The Drunkard's Walk)")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
print("*"*50)
# distance from the start point to end.
final_distance = np.sqrt(x[-1]**2 + y[-1]**2) 
# Theoretical Expectation:
# According to Einstein's theory of Brownian Motion (Diffusion),
# the displacement scales with the square root of the number of steps.
expected_distance = np.sqrt(steps)

print(f"Actual Distance:   {final_distance:.2f} meters")
print(f"Expected Theory:   {expected_distance:.2f} meters")
```
